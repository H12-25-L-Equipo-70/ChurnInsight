```python
import pandas as pd
import requests

# Configure pandas to display all rows
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

url = "https://raw.githubusercontent.com/H12-25-L-Equipo-70/ChurnInsight/refs/heads/Data/dataset_empresas_fintech_v2.7.json"
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)
# Mostrar las filas donde la columna 'Churn' es TRUE
display(df[df['Churn'] == True])
```
